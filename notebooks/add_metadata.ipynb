```python
import pandas as pd
import json

print("hello")
```

```python
attribs = pd.read_json("../data/processed/movie_attributes.json", orient="index")
attribs.reset_index(inplace=True)
attribs.columns = ["title", "attributes"]
attribs
```

```python
movies = pd.read_csv('../data/processed/movie_and_plot.csv')
movies.head()
```

```python
df = pd.merge(movies, attribs, on="title", how="inner")
df
```

```python
df.drop(columns = ['plot'], inplace=True)
data_list = df.to_dict(orient='records')
json_data = {item['title']: {key: value for key, value in item.items() if key != 'title'} for item in data_list}
json_data
```

```python
with open("../data/processed/final_data.json", "w") as f:
    json.dump(json_data, f)
```
