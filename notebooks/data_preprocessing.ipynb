```python
import pandas as pd

movies = pd.read_csv("../data/raw/title.basics.tsv.gz", sep="\t", low_memory=False, header=0)
movies.head()
```

```python
movies.drop(columns=['originalTitle', 'isAdult', 'endYear', 'runtimeMinutes'], inplace=True)
# movies.rename(columns={'tconst': 'movie_id', 'primaryTitle': 'title', 'startYear': 'year'}, inplace=True)
movies.head()
```

```python
ratings = pd.read_csv("../data/raw/title.ratings.tsv.gz", sep="\t")
ratings.head()
```

```python
df = pd.merge(movies, ratings, on='tconst', how='inner')
df.head()
```

## sort all movies by rating

```python
df_movies = df[df['titleType'] == 'movie']
df_movies.reset_index(inplace=True, drop=True)
print(len(df_movies))
df_movies.head()
```

```python
df = df_movies.sort_values(by=['numVotes', 'averageRating'], ascending=False).head(1000)
df.reset_index(drop=True, inplace=True)
df
```

```python
df['tconst'] = df['tconst'].str.replace('tt', '').astype(int)
df.head()
```

```python
import json

with open('../data/processed/plot_data.json', 'r') as f:
    plot_data = json.load(f)

print(plot_data)
```

```python
plots = pd.read_json('../data/processed/plot_data.json', orient='index')
plots.reset_index(inplace=True)
plots.columns = ['tconst', 'plot']
plots.head()
```

```python
df = pd.merge(df, plots, on='tconst', how='inner')
df.head()
```
